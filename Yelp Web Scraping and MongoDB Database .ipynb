```python
info_list=[]
website_list=[]
address_list=[]
phone_list=[]
for a in range(1,41):
    filename = 'sf_donut_shop_'+str(a) +".html" 
    store_info={}
    ### read 40 files 
    with open(filename, 'r', encoding='utf-8') as f:
        webpage = BeautifulSoup(f.read(), 'html.parser') 
        info = webpage.find_all('div', attrs={'class': 'css-xp8w2v padding-t2__09f24__Y6duA padding-r2__09f24__ByXi4 padding-b2__09f24__F0z5y padding-l2__09f24__kf_t_ border--top__09f24__exYYb border--right__09f24__X7Tln border--bottom__09f24___mg5X border--left__09f24__DMOkM border-radius--regular__09f24__MLlCO background-color--white__09f24__ulvSM'})
        for j in info:
                ### address
                try:
                    items = j.find_all('p', attrs={'class':'css-qyp8bo'})
                    for i in items:
                        address=i.get_text(separator=";")
                        address_list.append(address)
                        k1="address"
                        v1=address
                        store_info[k1] = v1

                except:
                    print("No Address")
                    
                ### Phone    
                try:
                    items_1 = str(j.find_all('p', attrs={'class':'css-1p9ibgf'}))
                    phone=re.findall(r".*(\(?\d{3}\D{0,3}\d{3}\D{0,3}\d{4}).*?",items_1)[0]
                    phone="("+phone
                    phone_list.append("("+phone)
                    k2="phone"
                    v2=phone
                    store_info[k2] = v2
                except:
                    pass
                    
                ### website    
                try:
                    items_2 = j.find_all('p', attrs={'class':'css-1p9ibgf'})
                    website=re.findall(r"(https?://[^\s]+)",items_1)[0]
                    website=website.split("<")[0]
                    website_list.append(website)
                    k3="website"
                    v3=website
                    store_info[k3] = v3
                except:
                    pass
                
    info_list.append(store_info)             
  
```

### Update the collection in MongoDB

```python
import json
from pymongo import MongoClient 
  
# Making Connection
myclient = MongoClient("mongodb://localhost:27017/") 
   
# database 
db = myclient["Final_Yelp"]
   
website=[]
for i in range(0,40):
    try:
        website.append(final_list[i]["website"])
    except:
        website.append('')
address=[]
for i in range(0,40):
    try:
        address.append(final_list[i]["address"])
    except:
        address.append('')
phone=[]
for i in range(0,40):
    try:
        phone.append(final_list[i]["phone"])
    except:
        phone.append('')

        ### set index
db.sf_donut_shops.create_index('rank', unique = True)
### update the dataset
for i in range(0,40):
    db.sf_donut_shops.update_one(
        {"ranks":final_list[i]["rank"]},
        {'$set': {"Address":address[i],"Phone":phone[i], "Website":website[i]}})
```
